```python
df_idade_coletores = pd.read_csv("Dados_tratados_-_Coletores_2026.xlsx - Geral.csv", usecols=[2])
media = df_idade_coletores.mean()
mediana = df_idade_coletores.median()
print(f"Media de idades {media}\nMediana de idades {mediana}")
```

```python
import matplotlib.pyplot as plt

plt.hist(df_idade_coletores, bins=30, color="green", edgecolor='black')

plt.title("Histograma de idades")
plt.xlabel("Idades")
plt.ylabel("Frequência")

plt.show()
```
